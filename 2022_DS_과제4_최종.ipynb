```python
# 문제 5. 입력값을 이용하여 다항식 형태를 출력하는 함수(printPoly)와 다항식을 계산하는 함수(calcPoly)를 작성하라.
# 입력 예시: 2
# 출력 예시
# P(x) = +3x^4 -4x^3 +0x^2 +5x^1 +2x^0
# 28

def printPoly(p_x):
    term = len(p_x) - 1	# 최고차항 숫자 = 배열길이-1
    polyStr = "P(x) = "

    for i in range(len(p_x)):
        coef = p_x[i]	# 계수
        # 코드 작성 구간
        if coef >= 0 :
          polyStr += "+"
        polyStr += str(coef) + "x^" + str(term-i) + str(" ")
        # polyStr += str(coef) + "x^" + str(term) + str(" ")
        # term -= 1
        
    return polyStr


def calcPoly(xValue, p_x):
    retValue = 0
    term = len(p_x) - 1  # 최고차항 숫자 = 배열길이 - 1

    for i in range(len(p_x)):
        # 코드 작성 구간
        coef = p_x[i]
        retValue += coef * xValue ** term
        term -= 1

    return retValue


# 실행 구문 (아래 코드를 수정하지 마시오.)
px = [3, -4, 0, 5, 2]			# = +3x^4 -4x^3 +0x^2 +5x^1 +2x^0

pStr = printPoly(px)
print(pStr)

xValue = int(input("X 값-->"))

pxValue = calcPoly(xValue, px)
print(pxValue)
```

```python
# 문제 6. 2차원 입력값을 이용하여 다항식 형태를 출력하는 함수(printPoly)와 다항식을 계산하는 함수(calcPoly)를 작성하라.
# 입력 예시: 2
# 출력 예시
# P(x) = +7x^3 -4x^2 +5x^0
# 45

def printPoly(p_x):
    polyStr = "P(x) = "

    # 코드 작성 구간
    for i in range(len(p_x[0])):
      term = p_x[0][i]
      coef = p_x[1][i]
      if coef >= 0:
        polyStr += "+"
      polyStr += str(coef) + "x^" + str(term) + " "
    return polyStr   


def calcPoly(xValue, p_x):
    retValue = 0

    # 코드 작성 구간
    for i in range(len(p_x[0])):
      term = p_x[0][i]
      coef = p_x[1][i]

      retValue += coef * xValue ** term
    return retValue


# 실행 구문 (아래 코드를 수정하지 마시오.)
px = [ [3, 2, 0], [7, -4, 5] ]  # 1행: 차수, 2행: 계수

pStr = printPoly(px)
print(pStr)

xValue = int(input("X 값-->"))

pxValue = calcPoly(xValue, px)
print(pxValue)
```

```python
# 문제 7. 입력값을 이용하여 단순 연결 리스트를 생성하고 출력하라.
# 입력 예시
# dataArray = ["파이썬", "C", "C++", "JAVA", "MATLAB"]
# 출력 예시
# >> printNodes(head)
# 파이썬 C C++ JAVA MATLAB

class Node():
	def __init__ (self):
		self.data = None
		self.link = None

def printNodes(start):
      # 코드 작성 구간
  current = start # head를 current에 넣어주기
  if current == None: 
    return
  print(current.data, end=" ") # 첫번째 노드출력
  while current.link != None: 
    current = current.link # current(head)의 link가 현재노드로 연결되어있음
    print(current.data, end = " ")
  print()
# 왜 메모리 만듬? 메모리 = 컴퓨터상의 메모리 = 연결리스트 
    
# 실행 구문 (아래 코드를 수정하지 마시오.)
memory = [] # 메모리 
head, current, pre = None, None, None
dataArray = ["파이썬", "C", "C++", "JAVA", "MATLAB"]

node = Node()		# 첫 번째 노드
node.data = dataArray[0] # 첫 번째 노드데이터에 dataArray의 첫 번째 값 넣기
head = node # head에 첫 번째 노드 넣어주기
memory.append(node) # 메모리에 첫 번째 노드 추가


for data in dataArray[1:]:	# 두 번째 이후 노드
    pre = node # 이전 노드에 첫번째 노드 넣어주기
    node = Node() # 새로운 노드 (현재 노드) 만들기
    node.data = data # 현재노드의 데이터는 dataArray의 값
    pre.link = node # 이전 노드의 링크는 현재노드에 연결
    memory.append(node) # 메모리에 현재 노드 추가

# print(memory)
# <__main__.Node object at 0x7fc9b0ef4150>

printNodes(head)
```

```python
# 문제 8. 노드 삽입 함수를 추가하고 입력값을 이용하여 단순 연결 리스트를 생성하라.
# 출력 예시
# ## 초기 연결 리스트 ##
# 파이썬 C++ MATLAB
# ## C++ 위치에 C 추가 결과 ##
# 파이썬 C C++ MATLAB
# ## MATLAB 위치에 JAVA 추가 결과 ##
# 파이썬 C C++ JAVA MATLAB

class Node():
	def __init__ (self):
		self.data = None
		self.link = None

def printNodes(start):
    # 코드 작성 구간
    current = start
    if current == None:
      return
    print(current.data, end=" ")

    while current.link != None:
      current = current.link
      print(current.data, end= " ")
    print()

    
    


def insertNode(findData, insertData):
    global memory, head, current, pre

    # 코드 작성 부분(첫 번째에 노드 삽입)
    if head.data == findData:
      newnode = Node()
      newnode.data = insertData
      newnode.link = head
      head = newnode # newnode가 가장 앞에 있는 노드가 됨
    # 코드 작성 부분(중간에 노드 삽입)
    else:
      current = head
      find_flang = False
      while current.link != None:
        pre = current
        current = current.link # head는 current/ current는 pre / pre(current)의 link는 다음 노드, 
        if current.data == findData:
          newnode = Node()
          newnode.data = insertData
          newnode.link = current
          pre.link = newnode
          find_flag = True
          break
          
    # 코드 작성 부분(마지막에 노드 삽입)   
      if not find_flag: # flag 코드 사용할 때 이용
          newnode = Node()
          newnode.data = insertData
          current.link = newnode


# 실행 구문 (아래 코드를 수정하지 마시오.)
memory = []
head, current, pre = None, None, None
dataArray = ["파이썬", "C++", "MATLAB"]

node = Node()			# 첫 번째 노드
node.data = dataArray[0]
head = node
memory.append(node)

for data in dataArray[1:]:		# 두 번째 노드부터
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)

print('## 초기 연결 리스트 ##')
printNodes(head)

print('\n## 파이썬 위치에 R 추가 결과(첫번째 노드 삽입) ##')
insertNode("파이썬", "R")
printNodes(head)

print('\n## C++ 위치에 C 추가 결과(중간 노드 삽입) ##')
insertNode("C++", "C")
printNodes(head)

print('\n## MATLAB 위치에 JAVA 추가 결과(마지막 노드 삽입) ##')
insertNode("MATLAB", "JAVA")
printNodes(head)
```

```python
# 문제 9. 노드 삭제 함수를 추가하고 입력값을 이용하여 단순 연결 리스트를 생성하라.
# 출력 예시
# ## 초기 연결 리스트 ##
# R 파이썬 C C++ JAVA MATLAB
# ## R 삭제 결과 ##
# 파이썬 C C++ JAVA MATLAB
# ## C++ 삭제 결과 ##
# 파이썬 C JAVA MATLAB
# ## MATLAB 삭제 결과 ##
# 파이썬 C JAVA

class Node():
	def __init__ (self):
		self.data = None
		self.link = None


def printNodes(start):
    # 코드 작성 구간
    current = start
    if current == None:
      return
    print(current.data, end=" ")

    while current.link != None:
      current = current.link
      print(current.data, end=" ")
    print()



def deleteNode(deleteData):
    global memory, head, current, pre

    # 코드 작성 구간
    if head.data == deleteData:
      current = head
      head = current.link
      del current
    else:
      current = head
      while current.link != None:
        pre = current
        current = current.link
        if current.data == deleteData:
          pre.link = current.link # pre 노드의 링크는 current의 다음 노드에 연결
          del(current.data)
          break


    

# 실행 구문 (아래 코드를 수정하지 마시오.)
memory = []
head, current, pre = None, None, None
dataArray = ['R', '파이썬', 'C', 'C++', 'JAVA', 'MATLAB']

node = Node()			# 첫 번째 노드
node.data = dataArray[0]
head = node
memory.append(node)

for data in dataArray[1:] :		# 두 번째 노드부터
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)

print('## 초기 연결 리스트 ##')
printNodes(head)

print('\n## R 삭제 결과 ##')
deleteNode("R")
printNodes(head)

print('\n## C++ 삭제 결과 ##')
deleteNode("C++")
printNodes(head)

print('\n## MATLAB 삭제 결과 ##')
deleteNode("MATLAB")
printNodes(head)
```

```python
# 과제 1. 2차원 입력값을 이용하여 다항식 형태를 출력하는 함수(printPoly)와 다항식을 계산하는 함수(calcPoly)를 작성하라.
        ## 단, 계수는 2차원 입력값(계수a, 계수b)의 곱한 결과를 그대로 출력한다.

# 입력 예시: 5
# 출력 예시
# P(x) = +7x^3 -4x^2 +0x^0 
# X 값 --> 5
# 775

def printPoly(p_x):
    polyStr = "P(x) = "

    # 코드 작성 부분
    for i in range(len(p_x[0])):
      term = p_x[0][i]
      coef = p_x[1][i]*p_x[2][i]
      if coef>=0:
        polyStr += "+"
      polyStr += str(coef) + "x^" + str(term) + " "

    return polyStr


def calcPoly(xValue, p_x):
    retValue = 0

    # 코드 작성 부분
    for i in range(len(p_x[0])):
      term = p_x[0][i]
      coef = p_x[1][i]*p_x[2][i]
      retValue += coef * xValue ** term   
    
    return retValue



# 실행 구문 (아래 코드를 수정하지 마시오.)
px = [ [3, 2, 0], [1, 1, 0], [7, -4, 5] ]  # 1행: 차수, 2행: 계수a, 2행: 계수b

pStr = printPoly(px)
print(pStr)

xValue = int(input("X 값 --> "))

pxValue = calcPoly(xValue, px)
print(pxValue)
```

```python
# 과제 2. 입력값을 이용하여 단순 연결 리스트를 생성하고 '홀수 번째 요소'를 출력하라.
        ## 단, index는 1부터 시작한다.

# 입력 예시
# dataArray = ["파이썬", "C", "C++", "JAVA", "MATLAB"]
# 출력 예시
# >> printNodes(head)
# 1: 파이썬, 3: C++, 5: MATLAB,

class Node():
	def __init__ (self):
		self.data = None
		self.link = None

def printNodes(start):
    # 코드 작성 부분
    current = start
    if current == None:
      return
    print("1: {0}, ".format(current.data),end=" ")
    odd_list = []
    while current.link != None:
      current = current.link
      odd_list.append(current.data)
    for num in range(1,len(odd_list),2):
      print("{0}: {1}, ".format(num+2,odd_list[num]), end=" ")
    




# 실행 구문 (아래 코드를 수정하지 마시오.)
memory = []
head, current, pre = None, None, None
dataArray = ["파이썬", "C", "C++", "JAVA", "MATLAB"]

node = Node()		# 첫 번째 노드
node.data = dataArray[0]
head = node
memory.append(node)

for data in dataArray[1:]:	# 두 번째 이후 노드
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)

printNodes(head)
```

```python
# 과제 3. 2차원의 데이터를 사용하는 단순 연결 리스트를 완성하라.
# 출력 예시
## 초기 연결 리스트 ##
# [1, '파이썬'] [2, 'C'], [3, 'C++'], [4, 'JAVA'], [5, 'MATLAB'], 
# ## R 삽입 결과 ##
# [0, 'R'] [1, '파이썬'], [2, 'C'], [3, 'C++'], [4, 'JAVA'], [5, 'MATLAB'], 
# ## JAVA 삭제 결과 ##
# [0, 'R'] [1, '파이썬'], [2, 'C'], [3, 'C++'], [5, 'MATLAB'], 
# ## MATLAB 검색 결과 ##
# [5, 'MATLAB']

class Node():
    def __init__ (self):
        self.data = None
        self.link = None


def printNodes(start):
    current = start
    if current == None:
        return
    print(current.data, end = ' ')
    while current.link != None:
        current = current.link
        print(current.data, end = ', ')


def insertNode(findData, insertData):
    global memory, head, current, pre
   
    # 코드 작성 부분

    if head.data[1] == findData:
      newnode = Node()
      newnode.link = head
      newnode.data = insertData
      head = newnode
    
    else:
      current = head
      find_flag = False
      while current.link != None:
        pre = current
        current = current.link
  
      
        if current.data[1] == findData:
          newnode = Node()
          newnode.link = current
          newnode.data = insertData
          pre.link = newnode
          find_flag = True
          break

      if not find_flag:
        newnode = Node()
        newnode.data = insertData
        current.link = newnode


def deleteNode(deleteData):
    global memory, head, current, pre

    # 코드 작성 부분
    if head.data[1] == deleteData:
      current = head
      head = current.link
      del current
    else:
      current = head
      while current.link != None:
        pre = current
        current = current.link
        if current.data[1] == deleteData:
          pre.link = current.link
          del current.data
          break


def findNode(findData):
    global memory, head, current, pre
    current = head

    # 코드 작성 부분
    if current.data[1] == findData:
      result = current
    else:
      find_flag = False
      while current.link != None:
        current = current.link
        if current.data[1] == findData:
          result = current
          find_flag = True
          break
      if not find_flag:
        result = Node()

    return result




# 실행 구문 (아래 코드를 수정하지 마시오.)
memory = []
head, current, pre = None, None, None
dataArray = [ [1, '파이썬'], [2, 'C'], [3, 'C++'], [4, 'JAVA'], [5, 'MATLAB'] ]  # 2차원 데이터

node = Node()			# 첫 번째 노드
node.data = dataArray[0]
head = node
memory.append(node)

for data in dataArray[1:] :		# 두 번째 노드부터
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)

print('## 초기 연결 리스트 ##')
printNodes(head)

print('\n## R 삽입 결과 ##')
insertNode('파이썬', [0, 'R'])
printNodes(head)

print('\n## JAVA 삭제 결과 ##')
deleteNode('JAVA')
printNodes(head)

print('\n## MATLAB 검색 결과 ##')
f_node = findNode('MATLAB')
print(f_node.data)
```
